```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import matplotlib.cm as cm
from sklearn.preprocessing import StandardScaler
# Import MNIST data
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets('/home/carnd/datasets/MNIST_data', one_hot=False)
```

```python
def weight_variable(shape, scope):
    return tf.get_variable(scope, shape, 
                           initializer=tf.contrib.layers.xavier_initializer())

def bias_variable(shape, scope):
    return tf.get_variable(scope, shape, initializer=tf.constant_initializer(0.0))

def parametric_relu(_x, scope):
    alphas = tf.get_variable(scope, _x.get_shape()[-1],
                       initializer=tf.constant_initializer(0.0),
                        dtype=tf.float32)
    pos = tf.nn.relu(_x)
    neg = alphas * (_x - abs(_x)) * 0.5
    return pos + neg

def cnn2d(X, conv_num_outputs, conv_ksize, scope='conv'):
    in_channels = X.get_shape().as_list()[3]
    with tf.variable_scope(scope):
        W_conv = weight_variable([conv_ksize, conv_ksize, in_channels, conv_num_outputs], 'conv_W')
        b_conv = bias_variable([conv_num_outputs], 'conv_b')
        conv = tf.nn.conv2d(X, W_conv, strides=[1, 1, 1, 1], padding='SAME')

        #h_conv = tf.nn.relu(conv + b_conv)
        h_conv = parametric_relu(conv + b_conv, 'alpha')
    return h_conv
    
def cnn2d_maxpool(X, conv_num_outputs, conv_ksize, scope='conv'):
    in_channels = X.get_shape().as_list()[3]
    with tf.variable_scope(scope):
        W_conv = weight_variable([conv_ksize, conv_ksize, in_channels, conv_num_outputs], 'conv_W')
        b_conv = bias_variable([conv_num_outputs], 'conv_b')
        conv = tf.nn.conv2d(X, W_conv, strides=[1, 1, 1, 1], padding='SAME')

        #h_conv = tf.nn.relu(conv + b_conv)
        h_conv = parametric_relu(conv + b_conv, 'alpha')
        h_pool = tf.nn.max_pool(h_conv, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME')
    return h_pool

def fully_conn(x, num_outputs, activation = False, scope='fc'):
    with tf.variable_scope(scope):
        size = x.get_shape().as_list()
        W = weight_variable([size[1], num_outputs], 'fc_W')
        b = bias_variable([num_outputs], 'fc_b')
        dense = tf.add(tf.matmul(x, W), b)
    
        if activation is True:
            dense = parametric_relu(dense, 'alpha')
            
    return dense
    
def flatten(x):
    size = x.get_shape().as_list()
    flat = tf.reshape(x, [-1, size[1]*size[2]*size[3]])
    return flat

def get_center_loss(features, labels, alpha, nrof_classes):
    nrof_features = features.get_shape()[1]
    centers = tf.get_variable('centers', [nrof_classes, nrof_features], dtype=tf.float32,
            initializer=tf.random_normal_initializer(stddev=5), trainable=False)
    labels = tf.reshape(labels, [-1])
    centers_batch = tf.gather(centers, labels)
    
    #update centers
    unique_label, unique_idx, unique_count = tf.unique_with_counts(labels)
    appear_times = tf.gather(unique_count, unique_idx)
    appear_times = tf.reshape(appear_times, [-1, 1])
    
    diff = centers_batch - features
    diff = diff / tf.cast((1 + appear_times), tf.float32)
    diff = alpha * diff
    centers = tf.scatter_sub(centers, labels, diff)
    
    # Calculate center loss
    loss = tf.nn.l2_loss(features - centers_batch)
    return loss, centers

def calc_center_loss(features, label, alfa, nrof_classes):
    """Center loss based on the paper "A Discriminative Feature Learning Approach for Deep Face Recognition"
       (http://ydwen.github.io/papers/WenECCV16.pdf)
    """
    nrof_features = features.get_shape()[1]
    centers = tf.get_variable('centers', [nrof_classes, nrof_features], dtype=tf.float32,
        initializer=tf.random_normal_initializer(), trainable=False)
    label = tf.reshape(label, [-1])
    centers_batch = tf.gather(centers, label)
    diff = (1 - alfa) * (centers_batch - features)
    centers = tf.scatter_sub(centers, label, diff)
    loss = tf.nn.l2_loss(features - centers_batch)
    return loss, centers

def calc_center_loss_auto(features, label, alfa, nrof_classes):
    """Center loss based on the paper "A Discriminative Feature Learning Approach for Deep Face Recognition"
       (http://ydwen.github.io/papers/WenECCV16.pdf)
    """
    nrof_features = features.get_shape()[1]
    centers = tf.get_variable('centers', [nrof_classes, nrof_features], dtype=tf.float32,
        initializer=tf.random_normal_initializer(), trainable=True)
    label = tf.reshape(label, [-1])
    centers_batch = tf.gather(centers, label)
    loss = tf.nn.l2_loss(features - centers_batch)
    return loss, centers
```

```python
tf.reset_default_graph()
xs = tf.placeholder(tf.float32, [None, 784])
ys = tf.placeholder(tf.int64, [None, 1])

ys_one_hot = tf.one_hot(ys, 10)
keep_prob = tf.placeholder(tf.float32)
alpha_ = tf.placeholder(tf.float32)


x_image = tf.reshape(xs, [-1,28,28,1])

conv1 = cnn2d(x_image, 32, 5, 'conv1_1') 
conv1 = cnn2d_maxpool(conv1, 32, 5, 'conv1_2') 

conv2 = cnn2d(conv1, 64, 5, 'conv2_1') 
conv2 = cnn2d_maxpool(conv2, 64, 5, 'conv2_2') 

conv3 = cnn2d(conv2, 64, 5, 'conv3_1') 
conv3 = cnn2d_maxpool(conv3, 64, 5, 'conv3_2') 

flat = flatten(conv3)

embed = fully_conn(flat, 2, True, 'fc1') 
alpha = 0.01
center_loss, centers = calc_center_loss_auto(embed, ys, alpha_, 10)
fc2 = fully_conn(embed, 10, False, 'out')

softmax_loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=ys_one_hot, logits=fc2))

loss = softmax_loss  +  1e-4 * center_loss 
#loss = softmax_loss  + 0.01*alpha_ * center_loss

train_op = tf.train.AdamOptimizer(3e-4).minimize(loss)

result = tf.argmax(fc2,1)
ground_truth = tf.reshape(ys, [-1])
correct_prediction = tf.equal(result, ground_truth)
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

init = tf.global_variables_initializer()
batch_size = 256
n_epochs = 20
with tf.Session() as sess:
    sess.run(init)
    batches = len(mnist.train.labels) // batch_size
    test_batch = [mnist.validation.images, mnist.validation.labels]
    for epoch in range(n_epochs):
        alpha = min(alpha+0.01, 0.5)
        for i in range(batches):
            train_batch = mnist.train.next_batch(batch_size)
            _ = sess.run(train_op, feed_dict = {xs: train_batch[0], ys: np.reshape(train_batch[1], [-1,1]), alpha_:alpha})

        print ('LOSS, Softmax_loss, Center_loss', sess.run([loss, softmax_loss, center_loss], feed_dict = {xs: test_batch[0], ys: np.reshape(test_batch[1], [-1,1]), alpha_:alpha}))
        print ('ACC@VALID:', sess.run(accuracy, feed_dict = {xs: test_batch[0], ys: np.reshape(test_batch[1], [-1,1]), alpha_:alpha}))
    _embed, _centers = sess.run([embed, centers], feed_dict = {xs: mnist.validation.images, ys: np.reshape(mnist.validation.labels, [-1,1]), alpha_:alpha} )
```
