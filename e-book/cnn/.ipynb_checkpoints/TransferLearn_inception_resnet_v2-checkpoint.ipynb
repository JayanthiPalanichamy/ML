# Inception-Resnet-v2 transfer learning

```python
import tensorflow as tf
import numpy as np
from inception_resnet_v2 import inception_resnet_v2, inception_resnet_v2_arg_scope
import os
import time
import tensorflow.contrib.slim as slim
import PIL.Image
import imagenet
import matplotlib.pyplot as plt
import inception_preprocessing
```

```python
checkpoints_dir='/Users/vijay/workspace/research/machineLearning/datasets/models/'
height = 299
width = 299
channels = 3
```

```python
img = PIL.Image.open('dog.jpg')
img = np.float32(img)
```

```python
plt.imshow(img/255.)
plt.show()
```

```python
import urllib

# Create graph
with tf.Graph().as_default():
    X = tf.placeholder(tf.float32, shape=[None, height-20, width-20, channels])
    with slim.arg_scope(inception_resnet_v2_arg_scope()):
        logits, end_points = inception_resnet_v2(X, num_classes=1001,is_training=False)
    predictions = end_points['Predictions']
    PreLogitsFlatten = end_points['PreLogitsFlatten']
    
    init_fn = slim.assign_from_checkpoint_fn(
        os.path.join(checkpoints_dir, 'inception_resnet_v2_2016_08_30.ckpt'),
        slim.get_model_variables('InceptionResnetV2'))

    image_string = open('dog.jpg', "rb").read() 
    image = tf.image.decode_jpeg(image_string, channels=3)
    processed_image = inception_preprocessing.preprocess_image(image, width, height, is_training=False)
    processed_image  = tf.expand_dims(processed_image, 0)
    
    # Execute graph
    with tf.Session() as sess:
        init_fn(sess)
        img1 = processed_image.eval()
        #predictions_val = predictions.eval(feed_dict={X: img1})
        predictions_val, PreLogits = sess.run([predictions,PreLogitsFlatten], feed_dict={X: img1})
        probabilities = predictions_val[0, 0:]
        sorted_inds = [i[0] for i in sorted(enumerate(-probabilities), key=lambda x:x[1])]

    names = imagenet.create_readable_names_for_imagenet_labels()
    for i in range(5):
        index = sorted_inds[i]
        print('Probability %0.2f%% => [%s]' % (probabilities[index], names[index]))



```
